## Chi-Square
The Chi-Squared score computed will be used to determine whether there is no significant relation between two variables (null hypothesis) or vice versa (alternative hypothesis).  
The p-value is set at 0.05. I shall filter for values whereby p > 0.001 otherwise there will be too many combinations.

```python
combinations = list(product(targets, df.columns, repeat=1))
combinations
```

```python
chi2_results = []
for combination in combinations:
  if combination[0] != combination[1]:
    chi2_results.append(
      (combination[0], combination[1], list(ss.chi2_contingency(pd.crosstab(df[combination[0]], df[combination[1]])))[1])
    )
    
chi2_results
```

```python
p_lessthan_5pct_largerpoint1pct = [chi2 for chi2 in chi2_results if chi2[2] < 0.05 and chi2[2] > 0.001]
p_lessthan_5pct_largerpoint1pct
```

```python
chi_test_output = pd.DataFrame(p_lessthan_5pct_largerpoint1pct, columns=["var1", "var2", "coeff"])
chi_test_output.pivot(index="var1", columns="var2", values="coeff")
```

```python
p_lessthan_5pct = [chi2 for chi2 in chi2_results if chi2[2] < 0.05]
len(p_lessthan_5pct)
```

```python
print(sorted(p_lessthan_5pct)[50])
```

As we can observe, there are 618 combinations which have a p-value of less than 0.05. To some extent, this statistical analysis doesn't make quite a lot of sense, and I have interpreted it to be due to the lack of data.
